```python
import tkinter as tk
from tkinter import ttk, filedialog
import json
```

```python
class App:

    def __init__(self, name : str):
        self.root = tk.Tk(name)
    
    def run(self):
        self.root.mainloop()
```

```python
def changed(keys, widget_rows, sections):
        
    rowid = keys[0]
    section_menu = widget_rows[rowid]["Section Menu"]
    section_drop = widget_rows[rowid]["Section Dropbox"]
    course_menu = widget_rows[rowid]["Course Menu"]
    
    section_menu.set("-")
    section_drop["menu"].delete(0, "end")

    try:
        for section in sections[course_menu.get().lower()].keys():
            section_drop["menu"].add_command(label=section.upper(), command=tk._setit(section_menu, section.upper()))
    except KeyError:
        pass


def check_input(event, lst):
    
    combo_box = event.widget
    value = event.widget.get()

    if value == '':
        combo_box['values'] = lst
    else:
        data = []
        for item in lst:
            if value.lower() in item.lower():
                data.append(item)

        combo_box['values'] = data

def courses_submitted(widget_rows):
    
    for key in widget_rows:
        row = widget_rows[key]
        print(f"Selected Course: {row['Course Menu'].get()}")
        print(f"Selected Section: {row['Section Menu'].get()}")
        print()

def parse_timetable():
    file = filedialog.askopenfilename(filetypes=[("Excel Files", "*.xlsx")])

    print(file)
```

```python
def get_courses(timetable):
    return [s.title() for s in timetable["timetable"].keys()]
```

```python
def main():
    global_pady = 15
    global_rowNo = 0
    root = tk.Tk(screenName="TimeTable Generator")
    root.geometry("800x600")
    
    frame = tk.Frame(root)
    frame.pack()

    tt_button = tk.Button(frame, text="Parse Timetable", command = parse_timetable)
    tt_button.grid(column=0, row=global_rowNo, pady=global_pady)
    global_rowNo += 1

    ####################################################


    timetable = json.load(open("timetable.json"))
    courses = get_courses(timetable)
    sections = timetable["timetable"]
    
    total_courses = 7
    
    #####################################################
    
    widget_rows = dict()
    
    for i in range(total_courses):
        course_menu = tk.StringVar(master=root)
        row_id = str(course_menu)
        widget_rows[row_id] = dict()
        widget_rows[row_id]["Course Menu"] = course_menu
        widget_rows[row_id]["Course Menu"].set("Select Course")

        widget_rows[row_id]["Course Dropbox"] = ttk.Combobox(master=frame, height=20, 
                                                       textvariable=widget_rows[row_id]["Course Menu"], values=courses)
        widget_rows[row_id]["Course Dropbox"].config(width=100)
        widget_rows[row_id]["Course Dropbox"].grid(column=0, row=0)

        widget_rows[row_id]["Section Menu"] = tk.StringVar(master=root)
        widget_rows[row_id]["Section Menu"].set("-")

        widget_rows[row_id]["Section Dropbox"] = tk.OptionMenu(frame, widget_rows[row_id]["Section Menu"], "-")

        widget_rows[row_id]["Course Dropbox"].bind("<KeyRelease>", lambda e : check_input(e, courses))

        course_menu.trace("w", lambda *_ : changed(_, widget_rows, sections))

        widget_rows[row_id]["Course Dropbox"].grid(column=0, row=global_rowNo, pady=global_pady)
        widget_rows[row_id]["Section Dropbox"].grid(column=1, row=global_rowNo, pady=global_pady)
        global_rowNo += 1
        
    submit_btn = tk.Button(frame, text="Submit", command = lambda : courses_submitted(widget_rows))
    submit_btn.grid(column=0, row=global_rowNo, pady=global_pady)
    global_rowNo += 1

    root.mainloop()
```

```python
if __name__ == "__main__":
    main()
```
